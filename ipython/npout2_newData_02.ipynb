##column independency

```python
dfco = df.corr()
indices = np.where(dfco > 0.7)
cols = [(dfco.index[x], dfco.columns[y]) for x, y in zip(*indices) if x != y and x < y]

for col in cols:
  if col[1] in df.columns: del df[col[1]]
```

##chi-squared test
>remove C47, C48, C49

```python
from scipy.stats import chi2_contingency

#coding
#df['C211_C'] = df['C211'].replace({name: n for n, name in enumerate(df['C211'].unique())})

def chi2_test(df, colA, colB='C4'):
    chi2, p, dof, expected = chi2_contingency(pd.crosstab(df[colA], df[colB]))
    print "p: %s, reject H0 (%s affects %s)" % (p, colA, colB) if p<0.05 else "%s does NOT affect %s" % (colA, colB)
    return p

for col in [df.columns[i] for i in xrange(len(df.columns)) if df.dtypes[i]=='object' and df.columns[i] not in ['C1', 'C2', 'C3', 'C4']]:
    if chi2_test(df, col)>=0.05: del df[col]
```

##anova test
>remove C65, C71, C72, C73, C86, C88, C105, C113, C117, C121, C134, C137, C146, C155, C163, C176, C179, C180, C183, C192, C213

```python
from scipy.stats import f_oneway

def anova_test(df, col):
    df_y = df.loc[df['C4']=='Y', col].fillna(0) # or dropna ?
    df_n = df.loc[df['C4']=='N', col].fillna(0) # or dropna ?
    fVal, p = f_oneway(df_y, df_n)
    print "p: %s, reject H0 (%s df_y's mean != df_n's mean)" % (p, col) if p<0.05 else "%s no difference" % (col)
    return p

for col in [df.columns[i] for i in xrange(len(df.columns)) if df.dtypes[i]!='object' and df.columns[i] not in ['C1', 'C2', 'C3', 'C4']]:
    if not anova_test(df, col)<0.05: del df[col]
```

##Preprocessing

```python
from sklearn.preprocessing import StandardScaler # RobustScaler, scale

dfc = df.loc[:, [c for c in df.columns if c not in ['C1', 'C2', 'C3', 'C4']]]
for col in dfc.columns:
  try:
    dfc[col].fillna(0, inplace=True)
    dfc[col] = StandardScaler().fit_transform(dfc[col]) #(df['DT1_DATA_RTD_DUR_L'] - df['DT1_DATA_RTD_DUR_L'].mean())/df['DT1_DATA_RTD_DUR_L'].std(ddof=0)
  except:
    #print col
    del dfc[col]

dfc.head()
```

##Feature Selection

```python
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.pipeline import Pipeline
from sklearn import svm

#coding
X, y = dfc, df['C4'].replace({name: n for n, name in enumerate(df['C4'].unique())})
anova_svm = Pipeline([('anova', SelectKBest(f_regression, k=20)), ('svc', svm.SVC(kernel='linear'))])
anova_svm.set_params(svc__C=0.1).fit(X, y)

print anova_svm.score(X, y)

filtered = [dfc.columns[i] for i in anova_svm.named_steps['anova'].get_support(indices=True)]
print filtered
```

##SVM Classifier
